````python
def completeion(prompt, model='gpt-3.5-turbo'):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model = model,
        messages = messages,
        temperature = 0 #known as degree of randomness
    )
    return response.choices[0].message['content']
````
